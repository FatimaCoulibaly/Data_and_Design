Data Dive I: NYC Greenthumb Community Garden

I choose the data set, "NYC Greenthumb Community Gardens." Some of the information it includes is the garden name, borough, size, and council district. This data set is from NYC Opendata. The Department of Parks and Recreation provided this source. I choose this set because I'm interested in nature. Gardens are one of the ways that a person can add beauty or creativity to a neighborhood.

Results:

Brooklyn (B) is the borough with the most community gardens. In total it has 226 gardens. The neighborhood, East New York, has the most community gardens in and out of Brooklyn. There are 45 community gardens in this area.

After Brooklyn, Manhattan (M) follows with 151 gardens, next to the Bronx (X) with 120 gardens, and Queens (Q) with 36 gardens.

Staten Island (R) is the borough with the least amount of gardens. It only has 4 gardens.
