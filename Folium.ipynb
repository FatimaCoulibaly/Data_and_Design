```python
folium.Map(
    location=[44.1126, -73.9235],
          zoom_start=12,
          tiles='Stamen Terrain'
          )

tooltip = 'Click me!'

folium.Marker([44.2915, -73.9856], popup='<i>High Peaks</i>', tooltip=tooltip).add_to(m)
folium.Marker([44.2866, -73.9845], popup='<b>Golden Arrow Lakeside</b>', tooltip=tooltip).add_to(m)

m
```

```python
m = folium.Map(
    location=[44.1126, -73.9235],
    zoom_start=12,
    tiles='Stamen Terrain'
)

folium.Marker(
    location=[44.2915, -73.9856],
    popup='High Peaks',
    icon=folium.Icon(icon='cloud')
).add_to(m)

folium.Marker(
    location=[44.2866, -73.9845],
    popup='Golden Arrow Lakeside',
    icon=folium.Icon(color='green')
).add_to(m)

folium.Marker(
    location=[44.2938, -73.9827],
    popup='Mirror Lake Inn',
    icon=folium.Icon(color='red', icon='info-sign')
).add_to(m)


m
```

```python
m = folium.Map(
    location=[40.8116, -73.9465],
    zoom_start=12,
    tiles='Stamen Terrain'
)

folium.Marker(
    location=[40.8044, -73.9438],
    popup='Marcus Garvey Park',
    icon=folium.Icon(icon='cloud')
).add_to(m)

folium.Marker(
    location=[40.8061, -73.9585],
    popup='Morningside Park',
    icon=folium.Icon(color='green')
).add_to(m)

folium.Marker(
    location=[40.8175, -73.9487],
    popup='St. Nicholas Park',
    icon=folium.Icon(color='red', icon='info-sign')
).add_to(m)


m
```

```python
m = folium.Map(
    location=[40.8116, -73.9465],
    tiles='Stamen Toner',
    zoom_start=13
)

folium.Circle(
    radius=100,
    location=[40.8044, -73.9438],
    popup='Marcus Garvey Park',
    color='crimson',
    fill=False,
).add_to(m)

folium.CircleMarker(
    location=[40.8175, -73.9487],
    radius=50,
    popup='St. Nicholas Park',
    color='#3186cc',
    fill=True,
    fill_color='#3186cc'
).add_to(m)


m
```

```python
m = folium.Map(
    location=[44.1126, -73.9235],
    tiles='Stamen Terrain',
    zoom_start=13
)

m.add_child(folium.LatLngPopup())


m
```

```python
m = folium.Map(
    location=[42.1934, -74.0632],
    tiles='Stamen Terrain',
    zoom_start=13
)

folium.Marker(
    [42.2468, -74.0300],
    popup='Crystal Brook'
).add_to(m)

m.add_child(folium.ClickForMarker(popup='Waypoint'))


m
```
