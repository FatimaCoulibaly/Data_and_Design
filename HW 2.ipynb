## Problem 3: Data

```python
 data = {'id': [1,2,3,4], 'name': ['Hennessy', 'Tino', 'Oden', 'Lulu'],
        'age': [12,5,11,6], 'speed': [4.2,6.9,5.5,4.5], 'color': ['red', 'white', 'yellow', 'red']}
df = pd.DataFrame.from_dict(data)
```

```python
df['name']
#this will return the names of the different wines in a series
```

```python
df['color'].value_counts()
#this will return the value counts (number of) different colors of wine
```

```python
df.describe()
#this will return the descriptive statistics of all numerical columns
```
